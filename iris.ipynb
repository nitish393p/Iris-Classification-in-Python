IMPORT **LIBRARIES**

```python
#import the all essential libraries for our projects
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

LOADIN DATASET


```python
columns = ['Sepal length','Sepal width','Petal length','Petal Width','Class_labels']
# Load the data
df = pd.read_csv('iris.data', names=columns)
df.head()
```

VISUALIZATION OF DATASET


```python
df.describe()
```

```python
#Visualize the whole dataset
sns.pairplot(df, hue='Class_labels' , vars=['Sepal length', 'Sepal width', 'Petal length', 'Petal Width'] ,diag_kind='hist')
plt.show()
```

SEPERATING INPUT COLUMNS ANS THE OUTPUT COLUMNS

```python
#Seperate Features and target
data = df.values
#Features
X = data[1:,0:4]
Y = data[1:,4]
print(X)
print(Y)
```

SPLITTING THE DATA INTO TRAINING AND TESTING

```python
# Split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print(y_test)
```

MODEL1: SUPPORT VECTOR MACHINE ALGORITHM

```python
# Support vector machine algorithm
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
prediction1 = model_svc.predict(X_test)
# Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)
for i in range(len(prediction1)):
  print(y_test[i], ' ', prediction1[i])
```

MODEL2: LOGISTIC REGRESSION

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train, y_train)
```

```python
prediction2 = model_LR.predict(X_test)
# Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2)*100)
for i in range(len(prediction1)):
  print(y_test[i], ' ', prediction1[i])
```
